```python
import numpy as np
import os
from typing import Any, Dict
from metavision_core.event_io import EventsIterator
import hdf5plugin
# Path to the Metavision HDF5 file
hdf5_file_path = "/mnt/SSD_1TB/kria_rec.hdf5"

os.environ["HDF5_PLUGIN_PATH"] = hdf5plugin.PLUGIN_PATH

# Create an EventsIterator for the HDF5 file
try:
    event_iterator = EventsIterator(hdf5_file_path)

    # Accumulate all event batches into a single list
    all_events = []
    for events in event_iterator:
        if events.size > 0:
            all_events.append(events)

    # Concatenate all batches into a single NumPy structured array
    if all_events:
        events_array = np.concatenate(all_events)
        print(f"Successfully read {events_array.size} events into a NumPy array.")
        print("Array fields:", events_array.dtype.names)
        
        # Example of accessing fields
        timestamps = events_array['t']
        print(f"First 10 timestamps: {timestamps[:10]}")
    else:
        print("No events found in the HDF5 file.")

except Exception as e:
    print(f"Error reading the HDF5 file: {e}")
```

```python
events_array = np.concatenate(all_events)
```

```python
out_path = os.path.splitext(hdf5_file_path)[0] + "_events.npy"
np.save(out_path, events_array)
print(f"Saved: {out_path} (shape={events_array.shape}, dtype={events_array.dtype})")
```
